```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
%matplotlib inline
```

```python
yeild_apples= [0.99, 0.91, 0.9911, 0.92, 0.934]
plt.plot(yeild_apples);
```

```python
years=[2019, 2016, 2018, 2015, 2013]
yeild_apples= [0.99, 0.91, 0.9111, 0.92, 0.934]
plt.plot(years, yeild_apples)
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)");
```

```python
years = range(2000, 2007)
apples=[0.9, 1.1, 2.1, 3.1, 4.1, 5.1, 6.1]
oranges=[6.1, 7.1, 8.1, 9.1, 10.1, 11.1, 12.1]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel("Year")
plt.ylabel("Yield tons per hectare")
```

### Chart Title and Legend

```python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel("Year")
plt.ylabel("Yield tons per hectare")

plt.title("Crop Yields in Kanto")
plt.legend(["Apples", "Oranges"]);
```

### Line Markers

```python
plt.plot(years, apples, marker="p")
plt.plot(years, oranges, marker="x")

plt.xlabel("Year")
plt.ylabel("Yield tons per hectare")

plt.title("Crop Yield in Kanto")
plt.legend(["Apples", "Oranges"]);
```

### Styling Lines and Markers

```python
plt.plot(years, apples, marker="s", c="b", ls="-", lw=2, ms=8, mew=2, mec="navy")
plt.plot(years, oranges, marker="o", c="r", ls="--", lw=3, ms=10, alpha=.5)

plt.xlabel("Year")
plt.ylabel("Yield tons per hectare")

plt.title("Crop Yield in Kanto")
plt.legend(["Apples", "Oranges"]);
```

```python
plt.plot(years, apples, "s-g")
plt.plot(years, oranges, "or")

plt.xlabel("Year")
plt.ylabel("Yield tons per hectare")

plt.title("Crop Yield in Kanto")
plt.legend(["Apples", "Oranges"]);
```

### Changing the Figure Size

```python
plt.figure(figsize=(1, 6))

plt.plot(years, oranges, "o--r")
plt.title("Yield Of Oranges tons per hectare")
```
