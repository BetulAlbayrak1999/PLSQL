```python
years = range(2000, 2007)
apples=[0.9, 1.1, 2.1, 3.1, 4.1, 5.1, 6.1]
oranges=[6.1, 7.1, 8.1, 9.1, 10.1, 11.1, 12.1]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel("Year")
plt.ylabel("Yield tons per hectare")
```

### Chart Title and Legend

```python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel("Year")
plt.ylabel("Yield tons per hectare")

plt.title("Crop Yields in Kanto")
plt.legend(["Apples", "Oranges"]);
```

### Line Markers

```python
plt.plot(years, apples, marker="p")
plt.plot(years, oranges, marker="x")

plt.xlabel("Year")
plt.ylabel("Yield tons per hectare")

plt.title("Crop Yield in Kanto")
plt.legend(["Apples", "Oranges"]);
```

### Styling Lines and Markers

```python
plt.plot(years, apples, marker="s", c="b", ls="-", lw=2, ms=8, mew=2, mec="navy")
plt.plot(years, oranges, marker="o", c="r", ls="--", lw=3, ms=10, alpha=.5)

plt.xlabel("Year")
plt.ylabel("Yield tons per hectare")

plt.title("Crop Yield in Kanto")
plt.legend(["Apples", "Oranges"]);
```

### Scatter Plot

```python
# scatter plot used to show points on two dimentional grid

flower_df=sns.load_dataset("iris")
flower_df
```

```python
flower_df.species.unique()
```

```python
plt.plot(flower_df.sepal_length, flower_df.sepal_width);
```

```python
sns.scatterplot(x=flower_df.sepal_length, y=flower_df.sepal_width);
```

### Adding Hues

```python
sns.scatterplot(x= flower_df.sepal_length, 
                y= flower_df.sepal_width,
                hue= flower_df.species, s=100);
```

```python
plt.figure(figsize=(12, 5))
plt.title("Sepal Dimentions")

sns.scatterplot(x= flower_df.sepal_length, 
                y= flower_df.sepal_width, 
                hue=flower_df.species,
                s=100);
```

### Histogram

```python
# Histogram represents the distribution of data by using single type of variable
flower_df.describe()
```

```python
plt.title("Distribution of Sepal Width")
plt.hist(flower_df.sepal_width)
```

Controlling the size and number of bins

```python
plt.hist(flower_df.sepal_width, bins=5)
```

```python
plt.hist(flower_df.sepal_width, bins=np.arange(2, 5, 0.25))
```

```python
plt.hist(flower_df.sepal_width, 
         bins=[1, 3, 4, 4.5])
```

### Multiple Histograms


```python
setosa_df=flower_df[flower_df.species=="setosa"]
versicolor_df = flower_df[flower_df.species=="versicolor"]
virginica_df=flower_df[flower_df.species=="virginica"]
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins= np.arange(2, 5, 0.5));
plt.hist(virginica_df.sepal_width, alpha=0.4, bins= np.arange(2, 7, 0.5));
plt.legend(["setosa", "versicolor", "virginica"])
```

```python
plt.title("Distribution of Sepal width")
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], bins=np.arange(2, 5, 0.5), 
          stacked=False)
plt.legend(["Setosa", "Versicolor", "Virginica"])
```

### Heatmap


```python
sns.load_dataset("flights")
```

```python
flights_df= sns.load_dataset("flights").pivot(index="month", columns="year", values="passengers")
flights_df
```

```python
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights_df)
```

```python
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights_df, fmt="d", annot=True, cmap="Blues");
```
